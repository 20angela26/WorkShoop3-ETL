```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import pickle 
import os 
```

```python
df= pd.read_csv('../Data/World_Happiness_Report.csv')
```

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)


print(f'Tamaño del conjunto de entrenamiento: {train_df.shape[0]}')
print(f'Tamaño del conjunto de prueba: {test_df.shape[0]}')
```

```python
formula = (
    'Q("Happiness Score") ~ Q("GDP per Capita") + Q("Social support") + '
    'Q("Healthy life expectancy") + Q("Freedom") + Q("Generosity") + '
    'Q("Perceptions of corruption") + C(Country) + Year'
)

model = smf.ols(formula=formula, data=train_df).fit()
```

```python

train_countries = model.model.data.frame["Country"].unique()


new_countries = [c for c in test_df["Country"].unique() if c not in train_countries]

if new_countries:
    print(" Advertencia: Se encontraron países no vistos en el entrenamiento:")
    print(new_countries)
    print("Estos países serán excluidos del conjunto de prueba para evitar errores.\n")
    

    test_df = test_df[test_df["Country"].isin(train_countries)]
else:
    print("Todos los países del conjunto de prueba ya estaban en el entrenamiento.\n")


y_true = test_df["Happiness Score"]
y_pred = model.predict(test_df)


r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))



print(f"R²   : {r2:.4f}")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
```

```python
model.summary()
```

```python
path_model = os.path.join("../Model", "modelo_felicidad.pkl")

with open(path_model, "wb") as f:
    pickle.dump(model, f)

print(f"\n Modelo guardado correctamente en: '{path_model}'")
```
